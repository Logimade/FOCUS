```python
from ultralytics import FastSAM

# Define an inference source
source = "image.jpg"

# Create a FastSAM model
model = FastSAM("FastSAM-s.pt")  # or FastSAM-x.pt
```

```python
results = model(source, texts="hands")

annot = results[0].plot()  # numpy image with masks drawn
plt.figure(figsize=(10, 7))
plt.imshow(annot)
plt.axis("off")
plt.show()
```
